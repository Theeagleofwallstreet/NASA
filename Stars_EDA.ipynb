```python
# Calculate summary statistics for the numerical columns
summary_stats = stars_df[['Temperature', 'L', 'R', 'A_M']].describe()
```

```python
import matplotlib.pyplot as plt

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(stars_df['Temperature'], bins=20)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of Temperature')

plt.subplot(2, 2, 2)
plt.hist(stars_df['L'], bins=20)
plt.xlabel('Luminosity')
plt.ylabel('Frequency')
plt.title('Distribution of Luminosity')

plt.subplot(2, 2, 3)
plt.hist(stars_df['R'], bins=20)
plt.xlabel('Radius')
plt.ylabel('Frequency')
plt.title('Distribution of Radius')

plt.subplot(2, 2, 4)
plt.hist(stars_df['A_M'], bins=20)
plt.xlabel('Apparent Magnitude')
plt.ylabel('Frequency')
plt.title('Distribution of Apparent Magnitude')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot
#exploring the relationship between numerical features,
plt.scatter(stars_df['Temperature'], stars_df['L'])
plt.xlabel('Temperature')
plt.ylabel('Luminosity')
plt.title('Temperature vs Luminosity')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
stars_df['Color'].value_counts().plot(kind='bar')
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Distribution of Color')

plt.subplot(1, 3, 2)
stars_df['Spectral_Class'].value_counts().plot(kind='bar')
plt.xlabel('Spectral Class')
plt.ylabel('Count')
plt.title('Distribution of Spectral Class')

plt.subplot(1, 3, 3)
stars_df['Type'].value_counts().plot(kind='bar')
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Distribution of Type')

plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = stars_df[['Temperature', 'L', 'R', 'A_M']].corr()
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/JKFX/Downloads/Stars_data.csv")

numerical_features = ['Temperature', 'L', 'R', 'A_M']

sns.pairplot(data[numerical_features])
plt.tight_layout()
plt.show()
```

```python
categorical_features = ['Color', 'Spectral_Class', 'Type']

for feature in categorical_features:
    feature_counts = data[feature].value_counts()
    feature_counts.plot(kind='bar')
    plt.title(f'{feature} Distribution')
    plt.xlabel(f'{feature}')
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

categorical_feature = 'Type'
numerical_feature = 'Temperature'

data.boxplot(column=numerical_feature, by=categorical_feature)
plt.title(f'{categorical_feature} vs {numerical_feature}')
plt.xlabel(categorical_feature)
plt.ylabel(numerical_feature)
plt.show()
```

```python
categorical_feature = 'Type'
numerical_feature = 'Temperature'

sns.barplot(x=categorical_feature, y=numerical_feature, data=data)
plt.title(f'{categorical_feature} vs {numerical_feature}')
plt.show()
```

```python
categorical_feature = 'Type'
numerical_feature = 'Temperature'

data.groupby(categorical_feature)[numerical_feature].mean().plot(kind='bar')
plt.title(f'{categorical_feature} vs {numerical_feature}')
plt.xlabel(categorical_feature)
plt.ylabel(numerical_feature)
plt.show()
```
